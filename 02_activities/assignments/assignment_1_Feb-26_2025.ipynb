```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



np.random.seed(613) # set seed
x = np.arange(50) # 0,1,...,49
y = np.random.randint(0,100,50) # 50 random integers with min = 0, max = 99
fig,ax = plt.subplots(figsize=(5,3)) # figure size
ax.scatter(x,y)
ax.bar(x,y)
ax.plot(x,y)
ax.hist(y)
fig,ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y)

#add font for visualization
font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace', 'color':'green', 'size':14}

#specify the name, font and position  for x,y
ax.set_title('Total growth over time',fontdict = font1, loc = 'right')
ax.set_ylabel('Total growth',fontdict = font2)
ax.set_xlabel('Years since start',fontdict = font2)
plt.show()
```

```python
#add more color to the graph
fig,ax = plt.subplots(figsize=(5,3)) 
ax.plot(x,y,
           marker = '*',
           color = '#7425b9',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

 #specify the name, font and position  for x,y features
ax.set_title('Total growth over time',fontdict = font1, loc = 'right')
ax.set_ylabel('Total growth',fontdict = font2)
ax.set_xlabel('Years since start',fontdict = font2)
plt.show()
```
