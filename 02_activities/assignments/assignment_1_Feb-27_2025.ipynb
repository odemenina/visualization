```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array (["Luffy", "Zorro", "Nami", "Usop", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

fig, (ax1, ax2) = plt.subplots(ncols = 2,
                               nrows = 1,
                               figsize = (9,4)) # more suitable  size (9) for 2 graphs

font1 = {'family':'sans-serif', 'color':'black', 'size':10}
font2 = {'family':'serif', 'color':'blue', 'size':8}

# Create the scatter plot
ax1.scatter(x1, y1)

# Create the bar plot
ax2.bar(x2, y2)


# Create the bar plot with each bar having a different color
colors = plt.cm.viridis(np.linspace(0, 1, len(x2)))
bars = ax2.bar(x2, y2, color= colors) 

ax1.set_title('Total distribution over time',fontdict = font1)
ax1.set_ylabel('Total distribution',fontdict = font2)
ax1.set_xlabel('Years',fontdict = font2)

ax2.set_title('Total bars over time',fontdict = font1)
ax2.set_ylabel('Name',fontdict = font2)
ax2.set_xlabel('Years',fontdict = font2)

#Display the numbers on top of the bars
for bar in bars:
    height = bar.get_height()
    ax2.annotate(f'{height}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Display the plots
plt.show()
```

```python
#Multiple visualizations on the same axes
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])
fig, ax = plt.subplots(figsize=(7,3))

# call both bar and plot element to the same axes(ax)
ax.bar(x,y1,color = 'indigo')
ax.plot(x,y2,color = 'red')

```

```python

#calculate standard deviation of out data
y2_sd = np.std(y2)

#plot out line as before
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color= "red")
```

```python
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')

#add error information
ax.errorbar(x, y2,
        yerr = y2_sd,  ## specieid that we a re plotting the vertical error bars
        fmt = "none") ## makes sure we are not plotting the actual data points, only the error
```

```python
#customizing error apparance
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')
ax.errorbar(x,
        y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor = "indigo",
        elinewidth = 4,
        capsize = 5,
        capthick = 4)
```

```python
fig,ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2, color = 'red')

# if we dont want to see error for every single point, we can specify intervals using errorevery
ax.errorbar(x,
        y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor = "indigo",
        elinewidth = 4,
        capsize = 6,
        capthick = 4,
        errorevery = 2) ## errorevery variats the interval
```
